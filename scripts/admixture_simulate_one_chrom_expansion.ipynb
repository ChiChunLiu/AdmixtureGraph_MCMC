```python
model = momi.DemographicModel(N_e=1e4, gen_time=29, muts_per_gen=1.25e-8)

model.add_leaf("S2")
model.add_leaf("Adm")
model.add_leaf("S1")
model.add_leaf("Outgroup")

# at 40kya Adm receive a 30% pulse from S1
model.move_lineages("Adm", "GhostS1", t=3.99e4, p=.3)
# at 40kya GhostS1 joins onto S1
model.move_lineages("GhostS1", "S1", t=4.01e4)

model.set_size("Adm", N=500, t=7e4)
# at 85 kya Adm joins onto S2
model.move_lineages("Adm", "S2", t=8.5e4)

# at 250 kya S2 joins onto S1
model.move_lineages("S2", "S1", t=2e5)
# at 500 kya S1 joins onto S0
model.move_lineages("S1", "Outgroup", t=5e5)
```

```python
%matplotlib inline

yticks = [1e4, 2.5e4, 5e4, 7.5e4, 1e5, 2.5e5, 5e5, 7.5e5]

fig = momi.DemographyPlot(
    model, ["S2", "Adm", "GhostS1", "S1", "Outgroup"],
    figsize=(6,8),
    major_yticks=yticks,
    linthreshy=1e5, pulse_color_bounds=(0,.25))
```

```python
# simulating Data
recoms_per_gen = 1.25e-8
bases_per_locus = int(5e6)
n_loci = 100
ploidy = 2  

# n_alleles per population (n_individuals = n_alleles / ploidy)
sampled_n_dict = {"Outgroup": 4,"S1":8, "S2":8, "Adm":8}

# create data directory if it doesn't exist
!mkdir -p data/

# simulate 20 "chromosomes", saving each in a separate vcf file
for chrom in range(1, n_loci+1):
      model.simulate_vcf(
            f"data/{chrom}",
            recoms_per_gen=recoms_per_gen,
            length=bases_per_locus,
            chrom_name=f"1",
            ploidy=ploidy,
            random_seed=1234+chrom,
            sampled_n_dict=sampled_n_dict,xa
            force=True)
```
